```python
def plot_two_rewards_with_stats(rewards1, rewards2, window_1=100, window_2 = 100, labels=('Series 1', 'Series 2')):
    """
    Dibuja la media móvil y la mediana móvil de dos series de recompensas.

    Parámetros:
    -----------
    rewards1 : list o np.ndarray
        Array con el refuerzo total obtenido en cada episodio de la primera serie.
    rewards2 : list o np.ndarray
        Array con el refuerzo total obtenido en cada episodio de la segunda serie.
    window : int
        Tamaño de la ventana para calcular la media y mediana móvil.
    labels : tuple de dos strings
        Etiquetas para cada serie (por defecto ('Series 1', 'Series 2')).
    """
    # Convertimos ambos arrays a pd.Series
    s1 = pd.Series(rewards1)
    s2 = pd.Series(rewards2)

    # Calculamos media y mediana móviles
    mean1   = s1.rolling(window_1).mean()
    median1 = s1.rolling(window_1).median()
    mean2   = s2.rolling(window_2).mean()
    median2 = s2.rolling(window_2).median()

    # Configuramos estilo
    sns.set(style="whitegrid")
    plt.figure(figsize=(12, 6))

    # Colores base: C0 y C1 para diferenciar las dos series
    # 1) Dibujamos las series crudas con transparencia
    plt.plot(rewards1, label=f"{labels[0]}", color="C0", alpha=0.2, linewidth=1)
    plt.plot(rewards2, label=f"{labels[1]}", color="C1", alpha=0.2, linewidth=1)

    # 2) Media móvil de cada serie (línea continua)
    plt.plot(mean1,   label=f"{labels[0]} Media móvil", color="C0", linewidth=1)
    plt.plot(mean2,   label=f"{labels[1]} Media móvil", color="C1", linewidth=1)

    # 3) Mediana móvil de cada serie (línea punteada)
    plt.plot(median1, label=f"{labels[0]} Mediana móvil", color="C0",
             linewidth=1, linestyle='dotted')
    plt.plot(median2, label=f"{labels[1]} Mediana móvil", color="C1",
             linewidth=1, linestyle='dotted')

    # Personalización final
    plt.title("Recompensa total por episodio", fontsize=16)
    plt.xlabel("Episodio", fontsize=14)
    plt.ylabel("Recompensa total", fontsize=14)
    plt.legend(loc="upper left", fontsize=11)
    plt.tight_layout()
    plt.show()
```

```python
from collections import defaultdict

def plot_dual_choice_vs_offer(measurements1, measurements2, labels=('FFNN', 'RNN'), rel_value_A=2.2):
    """
    Compara dos listas de measurements graficando el porcentaje de elección de B por oferta.
    
    Parámetros:
    - measurements1, measurements2: listas tipo [['B', 'A'], (6, 1), 'B']
    - labels: tupla de dos etiquetas para la leyenda
    - rel_value_A: valor relativo de A en unidades de B (default 2.2)
    """
    
    def compute_stats(measurements):
        stats = defaultdict(lambda: {'B_count': 0, 'total': 0, 'refuerzo_relativo': 0.0})
        for idx, item in enumerate(measurements):
            if not isinstance(item, (list, tuple)) or len(item) != 3:
                continue
            pair, amounts, choice = item
            if not isinstance(amounts, (list, tuple)) or len(amounts) != 2:
                continue
            b, a = amounts
            try:
                key = f"{b}B:{a}A"
                val_A = a * rel_value_A
                val_B = b
                refuerzo = val_A / val_B if val_B != 0 else float('inf')
                stats[key]['total'] += 1
                if choice == 'B':
                    stats[key]['B_count'] += 1
                stats[key]['refuerzo_relativo'] = refuerzo
            except Exception as e:
                print(f"Error en entrada {idx}: {item} → {e}")
                continue
        return stats

    # Calcular estadísticas para ambas listas
    stats1 = compute_stats(measurements1)
    stats2 = compute_stats(measurements2)

    # Unificar todas las ofertas y sus refuerzos relativos
    all_offers = {}
    for offer, data in {**stats1, **stats2}.items():
        all_offers[offer] = data['refuerzo_relativo']

    # Ordenar las ofertas por refuerzo relativo descendente
    ofertas_ordenadas = sorted(all_offers.items(), key=lambda x: -x[1])

    x_labels = [offer for offer, _ in ofertas_ordenadas]

    # Extraer porcentajes para ambas listas
    y1, y2 = [], []
    for offer in x_labels:
        pct1 = (stats1[offer]['B_count'] / stats1[offer]['total'] * 100) if offer in stats1 else None
        pct2 = (stats2[offer]['B_count'] / stats2[offer]['total'] * 100) if offer in stats2 else None
        y1.append(pct1)
        y2.append(pct2)

    # Plot
    plt.figure(figsize=(10, 6))
    plt.scatter(x_labels, y1, color='C0', marker='o', s = 200, label=labels[0], alpha=0.3)
    plt.scatter(x_labels, y2, color='C1', marker='o', s = 200, label=labels[1], alpha = 0.3)
    plt.xlabel("Oferta (#B:#A)", fontsize=14)
    plt.ylabel("Porcentaje de elección de B", fontsize=14)
    plt.title("Curva psicométrica. 1A = 2.2B", fontsize=18)
    plt.ylim(-5, 105)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.legend(fontsize=14, loc='upper left', frameon=True)
    plt.tight_layout()
    plt.show()
```

```python
def activity_per_neuron(offers_list, firing_rates, first_steps=100, last_steps=100, n_rows=5, n_cols=10):
    """
    Plots mean firing rates per model unit for specified time windows across fixed-order offer types,
    using subplots for individual units, with individual Y-limits per subplot (min to max of that neuron's values).

    Parameters
    ----------
    offers_list : list
        List of length N_trials, each element like [[left_label, right_label], (left_drops, right_drops), choice],
        or None for invalid trials.
    firing_rates : np.ndarray
        Array of shape (n_units, max_timesteps, n_trials) with firing rates (NaN-padded for shorter episodes).
    first_steps : int, optional
        Number of time steps from episode start to average over (default=150).
    last_steps : int, optional
        Number of time steps from episode end to average over (default=150).
    n_rows : int, optional
        Number of subplot rows (default=5).
    n_cols : int, optional
        Number of subplot columns (default=10).

    Returns
    -------
    None
    """
    # Define the fixed offer order
    fixed_order = [(0, 1), (1, 2), (1, 1), (2, 1), (3, 1), (4, 1), (6, 1), (10, 1), (2, 0)]
    cat_labels = [f"{a}:{b}" for a, b in fixed_order]
    n_cats = len(fixed_order)

    n_units, max_ts, n_trials = firing_rates.shape

    # Filter valid trials and map to A,B counts
    valid_trials = []
    trial_ratios = []
    for i, entry in enumerate(offers_list):
        if entry is None:
            continue
        labels, counts, _ = entry
        left_label, right_label = labels
        left_cnt, right_cnt = counts
        # Determine A_cnt, B_cnt regardless of side
        if left_label == 'A':
            A_cnt, B_cnt = left_cnt, right_cnt
        else:
            A_cnt, B_cnt = right_cnt, left_cnt
        valid_trials.append(i)
        trial_ratios.append((A_cnt, B_cnt))

    # Preallocate result arrays
    fr_start = np.zeros((n_units, n_cats))
    fr_end = np.zeros((n_units, n_cats))

    # Compute means per fixed category
    for ci, ratio in enumerate(fixed_order):
        # find trials matching this ratio exactly
        trial_idxs = [ti for ti, r in zip(valid_trials, trial_ratios) if r == ratio]
        if not trial_idxs:
            continue
        # Mean for first window
        fr1 = np.nanmean(firing_rates[:, :first_steps, trial_idxs], axis=(1, 2))
        # Mean for last window
        fr2_trials = []
        for ti in trial_idxs:
            valid_ts = ~np.all(np.isnan(firing_rates[:, :, ti]), axis=0)
            if not np.any(valid_ts):
                continue
            last_idx = np.where(valid_ts)[0].max()
            start_idx = max(0, last_idx - last_steps + 1)
            fr2_trials.append(np.nanmean(firing_rates[:, start_idx:last_idx+1, ti], axis=1))
        fr2 = np.nanmean(np.stack(fr2_trials, axis=1), axis=1) if fr2_trials else np.zeros(n_units)
        # Store
        fr_start[:, ci] = fr1
        fr_end[:, ci] = fr2

    # Create subplots for all units
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(n_cols*2, n_rows*1.8), sharex=True, sharey=False)
    axes = axes.flatten()
    x = np.arange(n_cats)

    for u in range(n_units):
        ax = axes[u]
        y1 = fr_start[u]
        y2 = fr_end[u]
        ax.plot(x, y2, marker='o', linestyle='-', alpha=0.7, linewidth=0.8)
        ax.plot(x, y1, marker='o', linestyle='-', color='grey', alpha=0.7, linewidth=0.8)
        ax.set_title(f'Neuron {u+1}', fontsize=6)
        # Determine individual y-limits
        combined = np.concatenate([y1, y2])
        finite = combined[np.isfinite(combined)]
        if finite.size > 0:
            ymin, ymax = finite.min(), finite.max()
            if np.isclose(ymin, ymax):
                delta = ymin * 0.1 if ymin != 0 else 1
                ymin -= delta
                ymax += delta
            ax.set_ylim(ymin, ymax)
        # X axis ticks
        if u // n_cols == n_rows - 1:
            ax.set_xticks(x)
            ax.set_xticklabels(cat_labels, rotation=45, fontsize=6)
        else:
            ax.set_xticks([])
        # Y-axis label on first column
        if u % n_cols == 0:
            ax.set_ylabel('FR')

    # Remove unused axes
    for idx in range(n_units, len(axes)):
        fig.delaxes(axes[idx])

    plt.tight_layout()
    plt.suptitle('Firing Rates: primeros vs últimos time steps por neurona', y=1.02)
    plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

# Ensure warnings are managed, especially for nanmean
warnings.filterwarnings("ignore", category=RuntimeWarning)


def activity_per_selected_neurons_normalized_with_std(
    offers_list: list,
    firing_rates: np.ndarray,
    neuron_indices: list[int],
    first_steps: int = 100,
    last_steps: int = 100,
    y_padding_percent: float = 5.0
):
    """
    Plots MEAN NORMALIZED firing rates per specified neuron for defined time windows
    across fixed-order offer types, showing each neuron side-by-side.
    Firing rates are normalized per neuron using Min-Max scaling (0-1).
    Lines connect only the intermediate points (categories 1 to n-2). Vertical error bars indicate
    the standard deviation of the mean FR across trials for each point.
    Background grid is included.

    Parameters
    ----------
    offers_list : list
        List of length N_trials, each element like [[left_label, right_label], (left_drops, right_drops), choice],
        or None for invalid trials.
    firing_rates : np.ndarray
        Array of shape (n_units, max_timesteps, n_trials) with firing rates (NaN-padded for shorter episodes).
    neuron_indices : list[int]
        A list of 0-based indices of the neurons to plot.
    first_steps : int, optional
        Number of time steps from episode start to average over (default=100).
    last_steps : int, optional
        Number of time steps from episode end to average over (default=100).
    y_padding_percent : float, optional
        Percentage of the normalized range (0-1) to add as padding
        to the upper and lower limits of the Y-axis (default=5.0).

    Returns
    -------
    None
    """
    # Define the fixed offer order
    fixed_order = [(0, 1), (1, 2), (1, 1), (2, 1), (3, 1), (4, 1), (6, 1), (10, 1), (2, 0)]
    cat_labels = [f"{a}:{b}" for a, b in fixed_order]
    n_cats = len(fixed_order)

    # Lines connecting intermediate points require at least 3 points in the sequence (0, 1, 2)
    # If n_cats = 3, intermediate points are just index 1. Need >= 4 points for a line segment.
    # Let's define intermediate range as indices 1 to n_cats-2. This requires n_cats >= 3.
    # A line segment between intermediate points requires n_cats >= 4 for indices 1 and 2 to exist.
    # Revisiting the definition: "puntos de la izquierda del todo y el de la derecha del todo no se unan con líneas al resto de puntos (los del medio)".
    # This implies point 0 is not connected to 1, and point n_cats-1 is not connected to n_cats-2.
    # So the line segment should run from index 1 to index n_cats-2.
    # This is possible if indices 1 and n_cats-2 exist and are valid, requiring n_cats >= 3.
    # If n_cats == 3, indices are 0, 1, 2. Intermediate points are slice(1, 2) which is just index 1. No line segment possible.
    # If n_cats == 4, indices are 0, 1, 2, 3. Intermediate points are slice(1, 3) which are indices 1, 2. A line connects 1 and 2.
    # So, lines should connect points with indices 1, 2, ..., n_cats-2.
    # This requires n_cats >= 3 for the slice [1:n_cats-1] to be non-empty.
    # A line requires at least 2 points, so indices 1 and 2 must exist, meaning n_cats >= 4.
    # Let's refine the condition: Plot lines between points from index 1 to index n_cats-2 IF n_cats >= 3.
    # If n_cats == 3, indices are 0, 1, 2. Slice is [1:2], just index 1. No line.
    # If n_cats >= 4, indices are 0, 1, ..., n_cats-2, n_cats-1. Slice is [1:n_cats-1], indices 1, ..., n_cats-2. These can form lines.
    # Okay, the condition should be n_cats >= 3 to *define* the intermediate range [1:n_cats-1].
    # Plotting a line using a set of points will automatically draw segments between adjacent points in the set.
    # So we just need to provide the points from index 1 to n_cats-2.
    plot_intermediate_lines = n_cats >= 3 # Enable lines if there are at least 3 categories

    n_units, max_ts, n_trials = firing_rates.shape

    # --- Data preparation for ALL neurons ---
    valid_trials = []
    trial_ratios = []
    for i, entry in enumerate(offers_list):
        if entry is None:
            continue
        try:
            labels, counts, _ = entry
            left_label, right_label = labels
            left_cnt, right_cnt = counts

            if left_label == 'A':
                 A_cnt, B_cnt = left_cnt, right_cnt
            elif left_label == 'B':
                 A_cnt, B_cnt = right_cnt, left_cnt
            else:
                 continue # Skip if unexpected juice labels

            if (A_cnt, B_cnt) not in fixed_order:
                 continue # Skip if ratio not in fixed_order

            valid_trials.append(i)
            trial_ratios.append((A_cnt, B_cnt))
        except Exception as e:
            print(f"Warning: Error processing trial {i} measurements: {e}. Skipping trial.")
            continue


    fr_start_mean_raw = np.full((n_units, n_cats), np.nan)
    fr_end_mean_raw = np.full((n_units, n_cats), np.nan)
    fr_start_std_raw = np.full((n_units, n_cats), np.nan)
    fr_end_std_raw = np.full((n_units, n_cats), np.nan)


    for ci, ratio in enumerate(fixed_order):
        trial_idxs_for_ratio = [valid_trials[j] for j, r in enumerate(trial_ratios) if r == ratio]

        if not trial_idxs_for_ratio:
            continue

        fr_tensor_subset = firing_rates[:, :, trial_idxs_for_ratio]

        # --- First window ---
        fr_start_per_trial = np.nanmean(fr_tensor_subset[:, :first_steps, :], axis=1)
        fr_start_mean_raw[:, ci] = np.nanmean(fr_start_per_trial, axis=1)
        if fr_start_per_trial.shape[1] > 1:
            fr_start_std_raw[:, ci] = np.nanstd(fr_start_per_trial, axis=1)
        else:
            fr_start_std_raw[:, ci] = np.nan


        # --- Last window ---
        fr_end_per_trial_list = [[] for _ in range(n_units)]

        for trial_idx in trial_idxs_for_ratio:
            valid_ts_in_trial = np.where(~np.isnan(firing_rates[0, :, trial_idx]))[0]

            if not valid_ts_in_trial.size > 0:
                continue

            last_idx = valid_ts_in_trial.max()
            start_idx = max(0, last_idx - last_steps + 1)

            fr_last_window_this_trial = firing_rates[:, start_idx:last_idx+1, trial_idx]

            mean_fr_this_trial = np.nanmean(fr_last_window_this_trial, axis=1)

            for u in range(n_units):
                 if not np.isnan(mean_fr_this_trial[u]):
                     fr_end_per_trial_list[u].append(mean_fr_this_trial[u])

        for u in range(n_units):
            if fr_end_per_trial_list[u]:
                 fr_end_mean_raw[u, ci] = np.nanmean(fr_end_per_trial_list[u])
                 if len(fr_end_per_trial_list[u]) > 1:
                     fr_end_std_raw[u, ci] = np.nanstd(fr_end_per_trial_list[u])
                 else:
                     fr_end_std_raw[u, ci] = np.nan


    # --- Normalization (Min-Max 0-1) per neuron ---
    fr_start_mean_norm = np.full_like(fr_start_mean_raw, np.nan)
    fr_end_mean_norm = np.full_like(fr_end_mean_raw, np.nan)
    fr_start_std_norm = np.full_like(fr_start_std_raw, np.nan)
    fr_end_std_norm = np.full_like(fr_end_std_raw, np.nan)

    for u in range(n_units):
        y1_mean_raw = fr_start_mean_raw[u, :]
        y2_mean_raw = fr_end_mean_raw[u, :]
        combined_mean_raw = np.concatenate([y1_mean_raw, y2_mean_raw])

        min_val = np.nanmin(combined_mean_raw)
        max_val = np.nanmax(combined_mean_raw)

        data_range = max_val - min_val

        # Normalization for means
        if np.isnan(min_val) or np.isnan(max_val):
            pass
        elif np.isclose(min_val, max_val):
            finite_mask_y1 = ~np.isnan(y1_mean_raw)
            fr_start_mean_norm[u, finite_mask_y1] = 0.5
            finite_mask_y2 = ~np.isnan(y2_mean_raw)
            fr_end_mean_norm[u, finite_mask_y2] = 0.5
        else:
            fr_start_mean_norm[u, :] = (y1_mean_raw - min_val) / data_range
            fr_end_mean_norm[u, :] = (y2_mean_raw - min_val) / data_range

            # Normalization for STDs: std_norm = std_raw / range
            # Only normalize if data_range is not zero/NaN
            if data_range > 0 and not np.isnan(data_range):
                finite_mask_std1 = ~np.isnan(fr_start_std_raw[u, :])
                fr_start_std_norm[u, finite_mask_std1] = fr_start_std_raw[u, finite_mask_std1] / data_range

                finite_mask_std2 = ~np.isnan(fr_end_std_raw[u, :])
                fr_end_std_norm[u, finite_mask_std2] = fr_end_std_raw[u, finite_mask_std2] / data_range


    # --- Plotting for Selected Neurons ---
    num_selected_neurons = len(neuron_indices)
    if num_selected_neurons == 0:
        print("No neuron indices provided to plot.")
        return

    fig, axes = plt.subplots(1, num_selected_neurons, figsize=(num_selected_neurons*3.5, 4.5), sharex=True, sharey=False)
    if num_selected_neurons == 1:
        axes = [axes]

    x = np.arange(n_cats)

    # Calculate Y-axis limits with padding
    padding_amount = (y_padding_percent / 100.0)
    y_min_plot = 0.0 - padding_amount
    y_max_plot = 1.0 + padding_amount

    plotted_handles_labels = []
    first_valid_plot_done = False

    for i, u_idx_absolute in enumerate(neuron_indices):
        ax = axes[i]

        if u_idx_absolute < 0 or u_idx_absolute >= n_units:
            print(f"Warning: Invalid neuron index {u_idx_absolute} (must be 0-{n_units-1}). Skipping.")
            ax.set_title(f'Neurona {u_idx_absolute} (Inválida)', fontsize=8)
            ax.axis('off')
            continue

        y1_mean_norm = fr_start_mean_norm[u_idx_absolute]
        y2_mean_norm = fr_end_mean_norm[u_idx_absolute]
        y1_std_norm = fr_start_std_norm[u_idx_absolute]
        y2_std_norm = fr_end_std_norm[u_idx_absolute]

        # Filter out NaNs before plotting for points, lines, and error bars
        valid_mask1 = ~np.isnan(y1_mean_norm)
        valid_mask2 = ~np.isnan(y2_mean_norm)
        valid_std_mask1 = valid_mask1 & ~np.isnan(y1_std_norm)
        valid_std_mask2 = valid_mask2 & ~np.isnan(y2_std_norm)


        # --- Plotting points ---
        points1 = ax.plot(x[valid_mask1], y1_mean_norm[valid_mask1], marker='o', linestyle='None', alpha=0.7, color='grey', markersize=5)
        points2 = ax.plot(x[valid_mask2], y2_mean_norm[valid_mask2], marker='o', linestyle='None', alpha=0.7, color=sns.color_palette()[0], markersize=5)

        # --- Plotting error bars ---
        ax.errorbar(x[valid_std_mask1], y1_mean_norm[valid_std_mask1], yerr=y1_std_norm[valid_std_mask1],
                    fmt='None', ecolor='grey', capsize=3, alpha=0.7)
        ax.errorbar(x[valid_std_mask2], y2_mean_norm[valid_std_mask2], yerr=y2_std_norm[valid_std_mask2],
                    fmt='None', ecolor=sns.color_palette()[0], capsize=3, alpha=0.7)


        # --- Plotting lines connecting intermediate points (indices 1 to n_cats-2) ---
        if plot_intermediate_lines:
            intermediate_slice = slice(1, n_cats - 1) # Indices 1, 2, ..., n_cats-2

            # Filter points in the intermediate range that are valid
            x1_intermediate_valid = x[intermediate_slice][valid_mask1[intermediate_slice]]
            y1_intermediate_valid = y1_mean_norm[intermediate_slice][valid_mask1[intermediate_slice]]

            x2_intermediate_valid = x[intermediate_slice][valid_mask2[intermediate_slice]]
            y2_intermediate_valid = y2_mean_norm[intermediate_slice][valid_mask2[intermediate_slice]]

            # Plot lines using only the valid intermediate points.
            # Matplotlib connects points in the order they appear.
            # This automatically handles gaps if intermediate points are NaN.
            line1, = ax.plot(x1_intermediate_valid, y1_intermediate_valid, linestyle='-', color='grey', alpha=0.7, linewidth=1.5, marker='None') # marker='None' ensures line only
            line2, = ax.plot(x2_intermediate_valid, y2_intermediate_valid, linestyle='-', color=sns.color_palette()[0], alpha=0.7, linewidth=1.5, marker='None') # marker='None' ensures line only


        # Collect handles/labels for the single legend
        if not first_valid_plot_done:
            # Create dummy handles representing the lines and markers for the legend
            handle1 = plt.Line2D([], [], color='grey', marker='o', linestyle='-', markersize=5, linewidth=1.5)
            handle2 = plt.Line2D([], [], color=sns.color_palette()[0], marker='o', linestyle='-', markersize=5, linewidth=1.5)
            plotted_handles_labels = ((handle1, handle2), (f'Antes de la oferta', f'Durante la oferta'))
            first_valid_plot_done = True

        ax.set_title(f'Neurona {u_idx_absolute}', fontsize=10)

        # Set fixed Y-axis limits with padding
        ax.set_ylim(y_min_plot, y_max_plot)

        # Set X-axis ticks and labels
        ax.set_xticks(x)
        ax.set_xticklabels(cat_labels, rotation=45, ha='right', fontsize=7)

        # Set Y-axis label only for the first subplot
        if i == 0:
            ax.set_ylabel('Tasa de disparo normalizada', fontsize=14)
            ax.set_xlabel('Oferta (#B:#A)', fontsize=14)
            ax.set_ylim(-1.0, 3.0)
        else:
            ax.set_ylabel('')
            ax.set_xlabel('Oferta (#B:#A)', fontsize=14)
            ax.set_ylim(-0.2, 1.25)

        # --- Add grid ---
        ax.grid(True, linestyle='--', alpha=0.6)


    # Add a single legend for the entire figure
    if plotted_handles_labels:
        handles, labels = plotted_handles_labels
        fig.legend(handles, labels, loc='upper center', ncol=2, bbox_to_anchor=(0.5, -0.05), fontsize=14)

    plt.suptitle('Tasa de disparo normalizada en función de la oferta', y=1.0, fontsize=18)
    plt.show()
```

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import warnings # Import warnings to manage OLS output

# Define the mapping for regression indices to neuron types
TIPO_POR_REG_INDEX = {
    0: "chosen-action",
    1: "action-value-right",
    2: "action-value-left",
    3: "chosen-value-difference",
    4: "chosen-value-sum",
    5: "juice-position"
}

# Define the names of the regressors, including the constant
# These correspond to the columns in X_with_const
REGRESSOR_NAMES = ['const', 'Adiff', 'AVR', 'AVL', 'CV_diff', 'CV_sum', 'Jpos']


def classify_neurons_by_last_valid_with_tables(
    FR_tensor: np.ndarray,
    measurements: list,
    window_size: int,
    R_A: float,
    R_B: float,
    p_global_thresh: float = 0.01,
    min_trials_for_regression: int = 5 # Added parameter for clarity/control
) -> tuple[pd.DataFrame, pd.DataFrame, pd.DataFrame]:
    """
    Classifies neurons based on regression against behavioral variables using
    the mean firing rate in the last valid window. Returns three DataFrames:
    type, coefficients, and p-values, only for classified neurons.

    Args:
        FR_tensor: 3D numpy array (n_neurons x n_timesteps x n_episodes).
        measurements: List of trial data (same length as n_episodes).
                      Each element is typically a tuple or dict containing
                      (juice_pair_LR, offer_pair_BA, chosen_juice) or None
                      for invalid trials.
        window_size: The number of last valid time steps to average over.
        R_A: Value constant for juice A.
        R_B: Value constant for juice B.
        p_global_thresh: Global significance threshold for the OLS F-test.
                         Neurons must pass this test to be included in the output.
        min_trials_for_regression: Minimum number of valid trials with non-NaN
                                   firing rates in the window required to attempt
                                   regression for a neuron.

    Returns:
        A tuple of three pandas DataFrames:
        df_type: Contains 'neuron_idx' and 'type'.
        df_coefficients: Contains 'neuron_idx' and columns for each regressor's coefficient ('const', 'Adiff', etc.).
        df_pvalues: Contains 'neuron_idx' and columns for each regressor's p-value ('const', 'Adiff', etc.).
        These DataFrames only include rows for neurons that were successfully classified.
    """
    n_neurons, n_timesteps, n_trials = FR_tensor.shape

    if len(measurements) != n_trials:
        raise ValueError(f"measurements debe tener longitud igual a n_trials ({n_trials}), "
                         f"pero obtuvo {len(measurements)}")

    valid_trials_indices = [t for t in range(n_trials) if measurements[t] is not None]
    if len(valid_trials_indices) == 0:
        # Instead of raising error, return empty DataFrames gracefully
        print("Warning: No valid trials found in measurements. Returning empty DataFrames.")
        cols_coef = ['neuron_idx'] + REGRESSOR_NAMES
        cols_pval = ['neuron_idx'] + REGRESSOR_NAMES
        return pd.DataFrame(columns=['neuron_idx', 'type']), pd.DataFrame(columns=cols_coef), pd.DataFrame(columns=cols_pval)

    n_valid = len(valid_trials_indices)
    X_full = np.zeros((n_valid, 6))  # columns = [Adiff, AVR, AVL, CV_diff, CV_sum, Jpos]

    valid_x_rows_mask = np.ones(n_valid, dtype=bool) # Keep track of rows where X generation was successful

    for idx, t in enumerate(valid_trials_indices):
        try:
            juice_pair_LR, offer_pair_BA, chosen_juice = measurements[t]
            juice_L, juice_R = juice_pair_LR
            nB, nA = offer_pair_BA

            Jpos = 1 if juice_L == 'A' else 0

            if juice_L == 'A':
                AVL = nA * R_A
                AVR = nB * R_B
            else:
                AVL = nB * R_B
                AVR = nA * R_A

            if chosen_juice == juice_L:
                Adiff = -1
            elif chosen_juice == juice_R:
                Adiff = +1
            else:
                # Handle inconsistent chosen juice like in the time block function
                print(f"Warning: Trial {t}: chosen_juice (‘{chosen_juice}’) does not match juices offered. Marking trial as invalid for X matrix.")
                valid_x_rows_mask[idx] = False
                continue # Skip setting values for this row

            if Adiff == -1:
                CVL = AVL
                CVR = 0.0
            else:
                CVR = AVR
                CVL = 0.0

            X_full[idx, 0] = Adiff
            X_full[idx, 1] = AVR
            X_full[idx, 2] = AVL
            X_full[idx, 3] = CVR - CVL
            X_full[idx, 4] = CVR + CVL
            X_full[idx, 5] = Jpos

        except Exception as e:
             print(f"Warning: Error processing measurements for trial {t}: {e}. Marking trial as invalid for X matrix.")
             valid_x_rows_mask[idx] = False
             continue

    # Filter X_full and corresponding valid_trials_indices based on successful X generation
    X_full = X_full[valid_x_rows_mask, :]
    valid_trials_indices_processed = [valid_trials_indices[i] for i in range(n_valid) if valid_x_rows_mask[i]]
    n_valid_processed = len(valid_trials_indices_processed)

    if n_valid_processed < min_trials_for_regression:
         print(f"Warning: Not enough valid trials processed for regression after checking measurements ({n_valid_processed} < {min_trials_for_regression}). Returning empty DataFrames.")
         cols_coef = ['neuron_idx'] + REGRESSOR_NAMES
         cols_pval = ['neuron_idx'] + REGRESSOR_NAMES
         return pd.DataFrame(columns=['neuron_idx', 'type']), pd.DataFrame(columns=cols_coef), pd.DataFrame(columns=cols_pval)


    # Calculate mean firing rate for the last valid window for each *processed* valid trial
    FR_avg = np.full((n_neurons, n_valid_processed), np.nan)

    for idx_processed, t in enumerate(valid_trials_indices_processed):
        # Find valid indices *within* this trial
        trial_fr = FR_tensor[:, :, t] # Shape (n_neurons, n_timesteps)
        # Assuming validity is based on the first neuron's non-NaN values across time for simplicity
        # Or you might have a separate validity mask for time steps within a trial
        # Let's assume the original intent was to find time indices where FR is valid for *some* neuron
        # np.where(~np.isnan(FR_tensor[0, :, t]))[0] implies checking neuron 0's FR validity
        # A more general check might be np.where(~np.isnan(FR_tensor[:, :, t]).all(axis=0))[0] or similar
        # Sticking to original logic:
        valid_time_indices_in_trial = np.where(~np.isnan(FR_tensor[0, :, t]))[0]

        if valid_time_indices_in_trial.size == 0:
            # print(f"Warning: No valid timesteps found for neuron 0 in trial {t}. FR_avg will be NaN for this trial.")
            FR_avg[:, idx_processed] = np.nan # Set whole column to NaN for this trial
            continue # Move to next processed trial

        # Select last 'window_size' valid time steps
        if valid_time_indices_in_trial.size >= window_size:
            last_inds = valid_time_indices_in_trial[-window_size:]
        else:
            last_inds = valid_time_indices_in_trial # Use fewer if not enough valid steps

        # Calculate mean FR for the selected time steps for *all* neurons in this trial
        # This results in mean FR per neuron for this trial's last window
        # np.nanmean might be safer here too if some neurons have NaNs in the selected time window
        with warnings.catch_warnings(): # Suppress RuntimeWarning: Mean of empty slice
             warnings.filterwarnings("ignore", category=RuntimeWarning)
             FR_avg[:, idx_processed] = np.nanmean(FR_tensor[:, last_inds, t], axis=1)


    # Add constant to the filtered X matrix
    X_with_const = sm.add_constant(X_full, prepend=True)


    # Initialize lists to store records for the three tables
    type_records = []
    coef_records = []
    pval_records = []


    # Perform regression for each neuron
    for m in range(n_neurons):
        # Y is the vector of mean FRs for neuron m across the processed valid trials
        Y_m_all = FR_avg[m, :]
        mask_valid_Y = ~np.isnan(Y_m_all) # Mask based on non-NaN mean FR for this neuron/trial

        # Filter Y and X based on the mask
        Y_valid = Y_m_all[mask_valid_Y]
        X_valid = X_with_const[mask_valid_Y, :] # Select rows in X corresponding to non-NaN Y

        # Check if enough data points remain for regression *after* filtering NaNs in Y
        if Y_valid.shape[0] < min_trials_for_regression or X_valid.shape[0] < min_trials_for_regression:
            # Skip neurons without enough valid data points for regression in this window
            continue

        # Perform OLS regression
        try:
            # Suppress warnings from statsmodels (e.g., convergence, rank deficiency)
            with warnings.catch_warnings():
                warnings.filterwarnings("ignore")

                # Optional: Check for rank deficiency before fitting (can prevent errors)
                # rank_X = np.linalg.matrix_rank(X_valid)
                # if rank_X < X_valid.shape[1]:
                #      # print(f"Warning: X_valid matrix is rank deficient for neuron {m}. Skipping regression.")
                #      continue # Skip if design matrix is problematic

                model = sm.OLS(Y_valid, X_valid).fit()

            # --- Check Classification Criteria ---
            # A neuron is "classified" if the global F-test is significant
            # model.f_pvalue can be NaN if model fit failed numerically
            if not np.isnan(model.f_pvalue) and model.f_pvalue < p_global_thresh:

                # Determine neuron type based on min p-value among regressors (excluding constant)
                pvals_regressors = model.pvalues[1:] # p-values for regressors 0-5 (Adiff to Jpos)
                neuron_type = "unknown_type" # Default in case something goes wrong below

                # Find index of minimum p-value ignoring NaNs/Infs
                valid_pvals_indices = np.where(~np.isnan(pvals_regressors) & ~np.isinf(pvals_regressors))[0]

                if valid_pvals_indices.size > 0:
                     # Find the index within the filtered pvals array, then map back to original regressor index (0-5)
                     idx_min_p_relative = valid_pvals_indices[np.argmin(pvals_regressors[valid_pvals_indices])]
                     # The relative index directly corresponds to the keys in TIPO_POR_REG_INDEX (0 to 5)
                     neuron_type = TIPO_POR_REG_INDEX.get(idx_min_p_relative, "unknown_type_index")
                else:
                    # If no valid p-values among regressors, cannot determine specific type
                    # Although globally significant, this neuron's tuning is unclear w.r.t. individual vars
                    # Sticking to original logic: If global p < thresh, it *is* classified.
                    # The original argmin would likely return 0 or raise an error if all pvals were NaN.
                    # Let's fall back to a default type (e.g., index 0) if no valid min pval exists.
                     print(f"Warning: Neuron {m} passed global test but has no valid individual regressor p-values. Assigning type based on index 0.")
                     neuron_type = TIPO_POR_REG_INDEX.get(0, "chosen-action") # Fallback to index 0 type


                # --- Append records for this classified neuron ---

                # Table 1: Type
                type_records.append({
                    "neuron_idx": m,
                    "type": neuron_type
                })

                # Table 2: Coefficients
                coef_dict = {"neuron_idx": m}
                # model.params contains coefficients for [const, Adiff, AVR, AVL, CV_diff, CV_sum, Jpos]
                for name, coef_val in zip(REGRESSOR_NAMES, model.params):
                     coef_dict[name] = coef_val
                coef_records.append(coef_dict)

                # Table 3: P-values
                pval_dict = {"neuron_idx": m}
                # model.pvalues contains p-values for [const, Adiff, AVR, AVL, CV_diff, CV_sum, Jpos]
                for name, pval_val in zip(REGRESSOR_NAMES, model.pvalues):
                     pval_dict[name] = pval_val
                pval_records.append(pval_dict)

        except Exception as e:
            # Handle potential regression errors (e.g., perfect multicollinearity in X_valid)
            # Neurons causing errors are not classified, so they won't be added to the lists
            # print(f"Warning: Regression failed for neuron {m}: {e}. Skipping classification.")
            pass # Simply skip this neuron if regression fails


    # Convert lists to DataFrames
    df_type = pd.DataFrame(type_records)

    # Ensure all regressor columns exist in coefficient and p-value tables,
    # even if no neurons were classified (results in empty lists)
    coef_cols = ['neuron_idx'] + REGRESSOR_NAMES
    pval_cols = ['neuron_idx'] + REGRESSOR_NAMES

    df_coefficients = pd.DataFrame(coef_records, columns=coef_cols)
    df_pvalues = pd.DataFrame(pval_records, columns=pval_cols)


    return df_type, df_coefficients, df_pvalues
```

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from itertools import product

# Define the mapping for regression indices to neuron types
TIPO_POR_REG_INDEX = {
    0: "chosen-action",
    1: "action-value-right",
    2: "action-value-left",
    3: "chosen-value-difference",
    4: "chosen-value-sum",
    5: "juice-position"
}
# Keep UNCLASSIFIED_TYPE defined as it's used internally during classification
UNCLASSIFIED_TYPE = "unclassified"

def classify_neurons_by_time_blocks(
    FR_tensor: np.ndarray,
    measurements: list,
    R_A: float,
    R_B: float,
    first_period_end: int,
    block_size: int,
    p_global_thresh: float = 0.01,
    min_trials_for_regression: int = 5
) -> pd.DataFrame:
    """
    Classifies neurons by type based on regression against behavioral variables
    within specific time blocks and returns counts per classified type per block.
    Unclassified neurons are not included in the output DataFrame.

    Args:
        FR_tensor: 3D numpy array (n_neurons x n_timesteps x n_episodes).
        measurements: List of trial data (same length as n_episodes).
                      Each element is typically a tuple or dict containing
                      (juice_pair_LR, offer_pair_BA, chosen_juice) or None
                      for invalid trials.
        R_A: Value constant for juice A.
        R_B: Value constant for juice B.
        first_period_end: The end time step (exclusive) for the first period
                          (e.g., 200 for steps 0-199).
        block_size: The size of each time block in time steps.
        p_global_thresh: Global significance threshold for the OLS F-test.
                         Neurons must pass this test to be classified.
        min_trials_for_regression: Minimum number of valid trials with non-NaN
                                   firing rates in the block required to attempt
                                   regression for a neuron.

    Returns:
        A pandas DataFrame with columns:
        'time_block_start': The starting time step of the block.
        'time_block_end': The ending time step (inclusive) of the block.
        'period': 'first_period' or 'last_period'.
        'neuron_type': The classified type (excluding 'unclassified').
        'count': The number of neurons of that type in that block.
    """
    n_neurons, n_timesteps, n_trials = FR_tensor.shape

    if len(measurements) != n_trials:
        raise ValueError(f"measurements must have length equal to n_trials ({n_trials}), "
                         f"but got {len(measurements)}")

    first_period_end = min(max(0, first_period_end), n_timesteps)
    if first_period_end == 0:
         print("Warning: first_period_end resulted in 0. The 'first_period' period will be empty.")

    if block_size <= 0:
        raise ValueError("block_size must be positive.")

    # --- 1. Prepare behavioral data (X matrix) ---
    valid_trials_indices = [t for t in range(n_trials) if measurements[t] is not None]
    if len(valid_trials_indices) == 0:
        print("Warning: No valid trials found in measurements.")
        return pd.DataFrame(columns=['time_block_start', 'time_block_end', 'period', 'neuron_type', 'count'])

    n_valid_trials = len(valid_trials_indices)
    X_full = np.zeros((n_valid_trials, 6))
    valid_x_rows_mask = np.ones(n_valid_trials, dtype=bool)

    for idx, t in enumerate(valid_trials_indices):
        try:
            juice_pair_LR, offer_pair_BA, chosen_juice = measurements[t]
            juice_L, juice_R = juice_pair_LR
            nB, nA = offer_pair_BA

            Jpos = 1 if juice_L == 'A' else 0

            if juice_L == 'A':
                AVL = nA * R_A
                AVR = nB * R_B
            else:
                AVL = nB * R_B
                AVR = nA * R_A

            if chosen_juice == juice_L:
                Adiff = -1
            elif chosen_juice == juice_R:
                Adiff = +1
            else:
                print(f"Warning: Trial {t}: chosen_juice (‘{chosen_juice}’) does not match juices offered. Marking trial as invalid for X matrix.")
                valid_x_rows_mask[idx] = False
                continue

            if Adiff == -1:
                CVL = AVL
                CVR = 0.0
            else:
                CVR = AVR
                CVL = 0.0

            X_full[idx, 0] = Adiff
            X_full[idx, 1] = AVR
            X_full[idx, 2] = AVL
            X_full[idx, 3] = CVR - CVL
            X_full[idx, 4] = CVR + CVL
            X_full[idx, 5] = Jpos

        except Exception as e:
             print(f"Warning: Error processing measurements for trial {t}: {e}. Marking trial as invalid for X matrix.")
             valid_x_rows_mask[idx] = False
             continue

    X_full = X_full[valid_x_rows_mask, :]
    valid_trials_indices = [valid_trials_indices[i] for i in range(n_valid_trials) if valid_x_rows_mask[i]]
    n_valid_trials_processed = len(valid_trials_indices)

    if n_valid_trials_processed < min_trials_for_regression:
         print(f"Warning: Not enough valid trials processed for regression after checking measurements ({n_valid_trials_processed} < {min_trials_for_regression}). Cannot perform analysis.")
         return pd.DataFrame(columns=['time_block_start', 'time_block_end', 'period', 'neuron_type', 'count'])

    X_with_const_full = sm.add_constant(X_full, prepend=True)

    # --- 2. Define Time Periods and Blocks ---
    periods = {
        'first_period': (0, first_period_end),
        'last_period': (max(0, n_timesteps - 100), n_timesteps)
    }

    last_period_start = n_timesteps - 100
    if last_period_start < 0:
         print(f"Warning: Total timesteps ({n_timesteps}) is less than 100. 'last_period' is invalid.")
         del periods['last_period']
    elif last_period_start < first_period_end:
         print(f"Warning: 'last_period' starts at {last_period_start} and overlaps with 'first_period' ending at {first_period_end}. Adjusting 'last_period' start to {first_period_end}.")
         periods['last_period'] = (first_period_end, n_timesteps)
         if periods['last_period'][0] >= periods['last_period'][1]:
             print(f"Warning: Adjusted 'last_period' period ({periods['last_period']}) is invalid. Removing it.")
             del periods['last_period']

    all_block_results = [] # This will temporarily store counts including 'unclassified'

    # --- 3. Process Each Time Block ---
    for period_name, (period_start, period_end) in periods.items():
        block_starts = range(period_start, period_end, block_size)

        for t_start in block_starts:
            t_end = min(t_start + block_size, period_end)
            if t_start >= t_end:
                continue

            FR_block_avg = np.full((n_neurons, n_valid_trials_processed), np.nan)

            for valid_idx_processed, trial_idx in enumerate(valid_trials_indices):
                fr_slice = FR_tensor[:, t_start:t_end, trial_idx]
                with warnings.catch_warnings():
                    warnings.filterwarnings("ignore", category=RuntimeWarning)
                    FR_block_avg[:, valid_idx_processed] = np.nanmean(fr_slice, axis=1)

            # --- 4. Classify Neurons for the Current Block ---
            block_neuron_classifications = []

            for m in range(n_neurons):
                Y_m_all = FR_block_avg[m, :]
                mask_valid_Y = ~np.isnan(Y_m_all)

                Y_valid = Y_m_all[mask_valid_Y]
                X_valid = X_with_const_full[mask_valid_Y, :]

                if Y_valid.shape[0] < min_trials_for_regression or X_valid.shape[0] < min_trials_for_regression:
                    block_neuron_classifications.append((m, UNCLASSIFIED_TYPE)) # Still classify internally as unclassified
                    continue

                try:
                    with warnings.catch_warnings():
                        warnings.filterwarnings("ignore")
                        rank_X = np.linalg.matrix_rank(X_valid)
                        if rank_X < X_valid.shape[1]:
                             block_neuron_classifications.append((m, UNCLASSIFIED_TYPE))
                             continue

                        model = sm.OLS(Y_valid, X_valid).fit()

                    if not np.isnan(model.f_pvalue) and model.f_pvalue < p_global_thresh:
                        pvals = model.pvalues[1:]
                        if np.all(np.isnan(pvals)) or np.all(np.isinf(pvals)):
                            neuron_type = UNCLASSIFIED_TYPE
                        else:
                             valid_pvals_indices = np.where(~np.isnan(pvals) & ~np.isinf(pvals))[0]
                             if valid_pvals_indices.size > 0:
                                 idx_min_p_relative = valid_pvals_indices[np.argmin(pvals[valid_pvals_indices])]
                                 idx_min_p_absolute = idx_min_p_relative
                                 neuron_type = TIPO_POR_REG_INDEX.get(idx_min_p_absolute, UNCLASSIFIED_TYPE)
                             else:
                                 neuron_type = UNCLASSIFIED_TYPE

                        block_neuron_classifications.append((m, neuron_type))
                    else:
                        block_neuron_classifications.append((m, UNCLASSIFIED_TYPE))

                except Exception as e:
                    block_neuron_classifications.append((m, UNCLASSIFIED_TYPE))

            # --- 5. Count Types for the Current Block (includes 'unclassified' for now) ---
            type_counts = pd.DataFrame(block_neuron_classifications, columns=['neuron_idx', 'neuron_type'])
            if not type_counts.empty:
                 counts = type_counts['neuron_type'].value_counts().reset_index()
                 counts.columns = ['neuron_type', 'count']
                 counts['time_block_start'] = t_start
                 counts['time_block_end'] = t_end -1
                 counts['period'] = period_name
                 all_block_results.append(counts)
            else:
                 # Append a dummy row for 'unclassified' if no neurons were processed
                 all_block_results.append(pd.DataFrame({
                     'neuron_type': [UNCLASSIFIED_TYPE],
                     'count': [n_neurons],
                     'time_block_start': [t_start],
                     'time_block_end': [t_end -1],
                     'period': [period_name]
                 }))


    # --- 6. Consolidate Results and Filter 'unclassified' ---
    if not all_block_results:
         print("No blocks processed.")
         return pd.DataFrame(columns=['time_block_start', 'time_block_end', 'period', 'neuron_type', 'count'])

    # Concatenate results including 'unclassified'
    df_results_with_unclassified = pd.concat(all_block_results, ignore_index=True)

    # Filter out the 'unclassified' type *before* creating the final output structure
    df_classified_only = df_results_with_unclassified[
        df_results_with_unclassified['neuron_type'] != UNCLASSIFIED_TYPE
    ].copy() # Use copy to avoid SettingWithCopyWarning

    # If no neurons were classified at all in any block
    if df_classified_only.empty:
         print("No neurons were classified in any block based on the criteria.")
         # Return a structure with existing blocks/periods but 0 counts for classified types
         all_blocks = df_results_with_unclassified[['time_block_start', 'time_block_end', 'period']].drop_duplicates()
         all_types = list(TIPO_POR_REG_INDEX.values()) # Only classified types
         index_combinations_list = [(block_tuple.time_block_start, block_tuple.time_block_end, block_tuple.period, neuron_type)
                                    for block_tuple, neuron_type in product(all_blocks.itertuples(index=False), all_types)]
         df_final = pd.DataFrame(index_combinations_list, columns=['time_block_start', 'time_block_end', 'period', 'neuron_type'])
         df_final['count'] = 0 # All counts are zero
         df_final.sort_values(by=['period', 'time_block_start', 'neuron_type'], inplace=True)
         return df_final


    # Ensure all relevant blocks and *classified* types are represented, even if count is 0
    all_blocks_classified = df_classified_only[['time_block_start', 'time_block_end', 'period']].drop_duplicates()
    all_classified_types = list(TIPO_POR_REG_INDEX.values()) # Only classified types

    # Create a list of all possible (block, classified_type) tuples
    index_combinations_list_classified = [(block_tuple.time_block_start, block_tuple.time_block_end, block_tuple.period, neuron_type)
                                          for block_tuple, neuron_type in product(all_blocks_classified.itertuples(index=False), all_classified_types)]

    # Create a DataFrame template with all combinations of blocks and classified types
    df_template_classified = pd.DataFrame(index_combinations_list_classified, columns=['time_block_start', 'time_block_end', 'period', 'neuron_type'])

    # Group the classified-only results to sum counts per block/type
    df_grouped_classified = df_classified_only.groupby(['time_block_start', 'time_block_end', 'period', 'neuron_type'])['count'].sum().reset_index()

    # Merge the template with the actual counts, filling missing combinations with 0
    df_final = pd.merge(df_template_classified, df_grouped_classified,
                        on=['time_block_start', 'time_block_end', 'period', 'neuron_type'],
                        how='left').fillna({'count': 0})
    df_final['count'] = df_final['count'].astype(int)

    # Sort for consistent output order
    df_final.sort_values(by=['period', 'time_block_start', 'neuron_type'], inplace=True)


    return df_final


# --- Function to Plot Results (remains the same) ---

def plot_neuron_type_counts_over_time(df_counts: pd.DataFrame):
    """
    Plots the count of each neuron type over time blocks.
    Assumes df_counts does NOT contain 'unclassified' type.

    Args:
        df_counts: DataFrame returned by classify_neurons_by_time_blocks.
                   Expected to only contain classified neuron types.
    """
    if df_counts.empty:
        print("No data to plot.")
        return

    periods = df_counts['period'].unique()

    for period in periods:
        df_period = df_counts[df_counts['period'] == period].copy()

        if df_period.empty:
             print(f"No data for period: {period}")
             continue

        df_period.sort_values(by='time_block_start', inplace=True)

        plt.figure(figsize=(15, 6))
        # Use all classified types from the entire df_counts for consistent hue mapping across periods
        all_classified_types = df_counts['neuron_type'].unique()
        # Define a consistent palette based on all possible types
        palette = sns.color_palette("tab10", len(all_classified_types))
        type_palette = dict(zip(all_classified_types, palette))


        sns.lineplot(data=df_period, x='time_block_start', y='count', hue='neuron_type', palette=type_palette)
        period_title_map = {
            'first_period': 'Primeros 200ms (aprox)',
            'last_period': 'Últimos 100ms (aprox)'
        }
        title = f'Número de neuronas por tipo en bloques de tiempo ({period_title_map.get(period, period)})'
        plt.title(title)
        plt.xlabel('Inicio del bloque de tiempo (ms)')
        plt.ylabel('Número de neuronas')
        plt.grid(True)
        plt.legend(title='Tipo de neurona', bbox_to_anchor=(1.05, 1), loc='upper left') # Move legend outside
        plt.tight_layout() # Adjust layout to prevent legend overlap
        plt.show()
```
